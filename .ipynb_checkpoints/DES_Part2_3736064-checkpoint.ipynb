#### Task 2.8.3: Variance Calculation


The difference with using the population variance formula (divide by n-1) rather than actual variance formula (divide by n) to calculate the variance of a sample is that it is biased. It is biased and produces an underestimation of the true variance. This bias in the value of variance for small sample size is much higher. Hence to compensate, we divide by n-1 (compensate by decreasing n to n-1) and not n. 

The amount of bias is larger when the sample size is smaller.
