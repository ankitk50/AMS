##### 3.3.1: Seed for Random Number Generation

##### 3.3.2: Bins for histograms

##### 3.3.3: System Utilization

The above results show, how the actual system utilization/throughput behaves compared to that of the system's rho, i.e. offered traffic.
Even though each actual value depends on random number seed, a general trend can be observed after running the simulation with different seeds. 
* When rho is small, i.e. 0.01 and 0.5, the actual system utilization will be almost rho (slightly smaller or bigger is possible).
* When rho becomes larger and larger, the actual system utilization will tend to be smaller than rho. Specifically when rho is 0.9, the actual value is around 0.8. This is because the queue has a capacity, which leads to packet drops when consecutive packets arrive with short IATs.

There is a theoretical value out of the analytical form of waiting-loss queueing system (see separate slide).
When we simulate for an infinite simulation time, the actual value will converge to the theoretical value.

When choosing an unlimited number of buffer spaces, the theoretical utilization will be equal to the offered traffic rho.
If we run it for infinite time, the actual utilization will also converge to rho.
